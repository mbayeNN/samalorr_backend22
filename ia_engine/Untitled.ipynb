```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import joblib

# 1. Données d'entraînement (remplacez par votre vrai dataset)
data = pd.DataFrame({
    'Temp': [33, 42, 35, 41, 32], 
    'Semaines': [12, 32, 20, 36, 10],
    'HTA': [0, 1, 0, 1, 0],
    'Risque': ['Faible', 'Élevé', 'Modéré', 'Élevé', 'Faible']
})

X = data[['Temp', 'Semaines', 'HTA']]
y = data['Risque']

# 2. Entraînement
model = RandomForestClassifier()
model.fit(X, y)

# 3. SAUVEGARDE CRUCIALE (pour le pont Laravel-Python)
joblib.dump(model, 'model_risques.pkl')
print("Modèle sauvegardé avec succès.")
```
